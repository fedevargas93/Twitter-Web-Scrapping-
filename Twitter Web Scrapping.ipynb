```python

import requests
import pandas as pd
import time

# Twitter Bearer Token
bearer_token = "AAAAAAAAAAAAAAAAAAAAAO4KsQEAAAAAS3ix1tfQb8S636RVpgaCQ8rF3c8%3DK95iEO55ZOh0iMQu7M8LAz7fEVbTAjuGXGk90lFeDhLhtfVRi3"

# Define search terms in English
search_tweets_english = ['Puerto Rican emigration',
 'Hurricane Maria',
 'Puerto Rico earthquake 2021',
 'Puerto Rican diaspora',
 'Puerto Ricans in the USA',
 'Post-2014 migration',
 'Reasons for migration from Puerto Rico',
 'Life of Puerto Ricans in the USA',
 'No return to Puerto Rico',
 'Corruption in Puerto Rico',
 'Lack of opportunities in Puerto Rico',
 'Puerto Rican experiences in the USA',
 'Puerto Rican community in the USA',
 'Impact of hurricane on migration',
 'Consequences of earthquake on migration',
 'Testimonies of Puerto Ricans in the USA',
 'Puerto Rico vs USA',
 'Puerto Rican migration decisions',
 'Stories of Puerto Rican migration',
 'Living as a Puerto Rican in the USA',
 'Challenges for Puerto Ricans in the USA',
 'Economic crisis in Puerto Rico',
 'Personal stories of Puerto Ricans',
 'Political impact on migration',
 'Tales of Puerto Rican migrants',
 'Economic development in Puerto Rico',
 'Adaptation of Puerto Ricans in the USA',
 'Diaspora experiences of Puerto Ricans',
 'Puerto Rican culture in the USA',
 'Challenges for Puerto Ricans in the USA',
 'Cultural influence on migration',
 'Health and migration in Puerto Rico',
 'Puerto Rican families in the USA',
 'Social impact of migration',
 'Education and migration in Puerto Rico',
 'Job opportunities in Puerto Rico',
 'Dreams of Puerto Ricans in the USA',
 'Social media and migration',
 'Views of Puerto Ricans about the USA',
 'Puerto Rico and the United States',
 'Puerto Rican',
 'Success stories of Puerto Ricans in the USA']
        
         
        
# API endpoint
api_url = "https://api.twitter.com/2/tweets/search/recent"

# Headers for authentication
headers = {
    "Authorization": f"Bearer {bearer_token}",
    "Content-Type": "application/json"
}

def fetch_tweets(search_term, max_results=100):
    params = {
        "query": f"{search_term} -is:retweet",  # Exclude retweets
        "max_results": max_results,
        "tweet.fields": "id,text,author_id,created_at"
    }
    collected_tweets = []
    next_token = None
    total_fetched = 0

    while total_fetched < 1000:
        if next_token:
            params["next_token"] = next_token
        
        response = requests.get(api_url, headers=headers, params=params)
        
        if response.status_code == 200:
            json_response = response.json()
            if 'data' in json_response:
                tweets = json_response['data']
                collected_tweets.extend(tweets)
                total_fetched += len(tweets)
                print(f"Fetched {len(tweets)} new tweets for '{search_term}'. Total fetched: {total_fetched}.")
                
                next_token = json_response.get('meta', {}).get('next_token', None)
                if not next_token:  # Break the loop if there's no next token
                    break
            else:
                print(f"No more tweets found for '{search_term}'.")
                break
        else:
            print(f"Failed to fetch tweets: {response.status_code}")
            break

        time.sleep(90)  # Wait for 90 seconds before the next request to avoid hitting rate limits
    
    return collected_tweets

# Main loop to fetch tweets for each search term and save them
if __name__ == "__main__":
    for term in search_tweets_english:
        print(f"Starting search for: {term}")
        tweets = fetch_tweets(term)
        if tweets:
            new_tweets_df = pd.DataFrame(tweets)
            new_tweets_df['search_term'] = term  # Add the search term as a column
            
            # Save the DataFrame to CSV, appending new data without overwriting
            with open("tweets_df_final.csv", "a", encoding='utf-8-sig', newline='') as file:
                new_tweets_df.to_csv(file, header=file.tell()==0, index=False)
                print(f"Data saved to tweets_df_final.csv. Total new tweets for '{term}': {len(new_tweets_df)}.")
        else:
            print(f"No new tweets fetched for term '{term}'.")

        print("Search completed.")
```
